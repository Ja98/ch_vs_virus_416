```python
import pandas
from functools import reduce
```

```python
#Load age demographic data
municipality_data_path = '../data/age_demographics/municipality/'
```

```python
#Process the files for each age category 
def read_data(category):
    df = pandas.read_excel(municipality_data_path + 'population_age_%s.xlsx'%(category), skiprows =5, usecols = 'B:D', index_col = None, nrows = 6804, 
            names = ['municipality_name', 'num_%s'%(category), 'pc_%s'%category])
    return df

categories = ['0to19','20to39','40to64','65_plus','80_plus']

dfs = [read_data(x) for x in categories]

#merge dataframes into one
df = reduce(lambda x, y: pandas.merge(x, y, on = 'municipality_name'), dfs)

#remove unnecessary rows
df.truncate(after = 2221)
```
