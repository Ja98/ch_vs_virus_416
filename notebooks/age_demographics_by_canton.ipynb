```python
import pandas
from functools import reduce
```

```python
#Load age demographic data
canton_data_path = '../data/age_demographics/canton/'
```

```python
#Process the files for each age category 
def read_data(category):
    df = pandas.read_excel(canton_data_path + 'population_age_%s.xlsx'%(category), skiprows =5, usecols = 'B:D', index_col = None, nrows = 26, 
            names = ['canton_name', 'num_%s'%(category), 'pc_%s'%category])
    return df

categories = ['0_19','20_39','40_64','65_plus','80_plus']

dfs = [read_data(x) for x in categories]

#merge dataframes into one
df = reduce(lambda x, y: pandas.merge(x, y, on = 'canton_name'), dfs)
```
