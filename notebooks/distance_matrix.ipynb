```python
import pandas as pd
import googlemaps
import numpy as np
```

API key should be requested from https://console.developers.google.com/. It is not free.

```python
API_key = 'You_API_key'  # enter Google Maps API key
gmaps = googlemaps.Client(key=API_key)
```

```python
Hospitals=pd.read_csv(r'../data/hospital_resources/Hospital_resources_hanmin.csv', encoding='ISO-8859-1')
total_num=len(Hospitals['ID'])
dist_mat=np.zeros((len(Hospitals['Latitude']),len(Hospitals['Latitude'])))  # distance between hospitals in meters
dura_mat=np.zeros((len(Hospitals['Latitude']),len(Hospitals['Latitude'])))  # duration between hospitals in seconds
Hospitals
```

```python
# Loop through coordinates of hospitals and extract distance according to coordinates via google API. 
for orig_ind in range(0, 66):
    LatOrigin = Hospitals['Latitude'][orig_ind]  
    LongOrigin = Hospitals['Longitude'][orig_ind] 
    origins = (LatOrigin, LongOrigin)  # define origin hospital
    for dest_ind in range(orig_ind+1, 67):
        LatDest = Hospitals['Latitude'][dest_ind] 
        LongDest = Hospitals['Longitude'][dest_ind]  
        destination = (LatDest, LongDest)  # define destination hospital
        print(orig_ind, dest_ind)
        result = gmaps.distance_matrix(origins, destination)
        dist = result['rows'][0]['elements'][0]['distance']['value'] # gives distance in meters
        dura = result['rows'][0]['elements'][0]['duration']['value'] # gives driving duration in seconds
        dist_mat[orig_ind, dest_ind] = dist
        print(dist)
        dura_mat[orig_ind, dest_ind] = dura       
        print(dura)
```

```python
upper_half = dist_mat
lower_half = upper_half.transpose()
dist_mat = np.add(upper_half, lower_half)
print(dist_mat)
```

```python
upper_half = dura_mat
lower_half = upper_half.transpose()
dura_mat = np.add(upper_half, lower_half)
print(dura_mat)
```

```python
np.save(r'../data/hospital_resources/dist_mat', dist_mat)
np.save(r'../data/hospital_resources/dura_mat', dura_mat)
```
